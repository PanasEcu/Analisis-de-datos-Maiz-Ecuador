```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("usa.csv")
```

```python
data
```

```python
ecuador = data[data.Area == "Ecuador"]
```

```python
ecuador
```

```python
peru = data[data.Area == "Peru"]
```

```python
peru
```

```python
usa = data[data.Area == "Estados Unidos de America"]
```

```python
usa
```

```python
plt.plot(ecuador.Ano, ecuador.Valor)
plt.plot(peru.Ano, peru.Valor)
plt.plot(usa.Ano, usa.Valor)
plt.legend(["Ecuador", "Perú" ,"EE.UU" ])
plt.title("Precio del Maiz en Ecuador, Perú y EE.UU")
plt.xlabel("Año")
plt.ylabel("Valor en dólares")
plt.show()
```
